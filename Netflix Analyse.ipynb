### Data Analysis & Visualizations

```python
# Comparing number of Movies vs TV shows
sns.countplot(x="type", data=df, palette="viridis")
plt.title("Number of Movies vs TV Shows on Netflix")
plt.xlabel("")
plt.ylabel("Count")
plt.show()
```

By above Graph we understand that Netflix has more movies than TV shows.

```python
# Finding which countries have the most Netflix titles

top_countries = df["country"].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("Top 10 Countries with Most Titles on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

By above graph we understand that the most netflix content originates in USA.

```python
# Finding the most common content ratings (like TV-MA, TV-14, etc.)

top_ratings = df["rating"].value_counts().head(10)

sns.barplot(x=top_ratings.values, y=top_ratings.index, palette="viridis")
plt.title("Most Common Ratings on Netflix")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

By above graph we understand that the most common Netflix ratings are TV-MA & TV-14.

```python
# Distribution of Movie Durations (in minutes)

# Filter only movies with duration in minutes
mask = df["duration"].str.contains("min", na=False)
movie_durations = df.loc[mask, "duration"].str.replace(" min", "").astype(int)

plt.hist(movie_durations, bins=30, color="orange", edgecolor="black")
plt.title("Distribution of Movie Durations (Minutes)")
plt.xlabel("Duration (min)")
plt.ylabel("Number of Movies")
plt.show()
```

By the above graph we understand that the most Netflix Movies are around 90-120 minutues long.

```python
#Top 10 Genres

# Split, clean, and count genres
genre_counts = df['listed_in'].dropna().str.split(',').explode().str.strip().value_counts().head(10)

# Plot top 10 genres
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette="viridis")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

By the above graph we understand that the most popular genres are internatinal movies, dramas and comedies.
